```python
import numpy as np # import numpy 
```

```python
import matplotlib.pyplot as plt #impot matplotlib
```

```python
location , scale = 100, 1
```

```python
s = np.random.logistic(location, scale, 5000)
```

```python
count, bins, ignored = plt.hist(s, bins=50)
```

```python
def logist(x, location, scale,): 
     return np.exp((location-x)/scale)/(scale*(1+np.exp((location-x)/scale))**2)
```

```python
 plt.plot(bins, logist(bins, location, scale)*count.max()/logist(bins, location, scale).max())
```

```python
plt.show()
```
